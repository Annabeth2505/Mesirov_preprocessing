```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def read_gct(filepath, summary= False): 
    df= pd.read_csv(filepath, delimiter= '\t', skiprows=2 )
    
    df= df.set_index([df.columns[0], df.columns[1]])
    if summary :
        print (f" Imported rows= {df.shape[0]} and columns = {df.shape[1]}")
        
    return df
```

```python
read_gct('BRCA_minimal_60x19 (1).gct', summary=True)
```
